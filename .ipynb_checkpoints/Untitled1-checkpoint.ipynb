```python
#MATPLOTLIB librería para la generacionde graficos a partir de datos contenidos en listas o arrays. Extensión: NumPy
```

```python
import matplotlib.pyplot as plt
```

```python
 #guarda las graficas
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,100,101)
x
```

```python
y = np.sqrt(x)+5
y
```

```python
plt.plot(x, y, color = 'green')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Primer gráfico')
plt.show()
```

```python
#Función subplot: Varios plots en el mismo canvas
#plt.subplot(filas, columnas, posicion)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color ='blue')
plt.subplot(1,2,2)
plt.plot(y,x, color='red')
plt.show()
```

```python
#GRÁFICOS CON EL MÉTODO ORIENTADO A OBJETOS
```

```python
grafico = plt.figure()
ejes = grafico.add_axes([0, 0, 1, 1])
ejes.plot(x,y)
ejes.set_xlabel('Eje x')
ejes.set_ylabel('Eje y')
plt.show()
```

```python
graficos2 = plt.figure()
ejes1 = graficos2.add_axes([0,0,1,1])
ejes2 = graficos2.add_axes([0.4,0.1,0.4,0.4])
ejes1.plot(x,y)
ejes2.plot(y,x)
ejes1.set_xlabel('raiz de x')
ejes2.set_xlabel('cuadrado de y')
plt.show()
```

```python
graficos,ejes=plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
graficos, ejes=plt.subplots(nrows=1, ncols=2)
for este_eje in ejes:
    este_eje.plot(x,y)
```

```python
graficos, ejes=plt.subplots(nrows=1, ncols=2)
ejes[0].plot(x,y)
ejes[1].plot(y,x)
plt.tight_layout
plt.show()
```

```python
#TAMAÑO Y RESOLUCION DE IMAGEN
```

```python
grafico = plt.figure(figsize=(3,4), dpi=100)
ejes = grafico.add_axes([0,0,1,1])
ejes.plot(x,y)
plt.show()
```

```python
graficos, ejes = plt.subplots(nrows=1, ncols=2, figsize=(4,3), dpi=100)
ejes[0].plot(x,y)
ejes[1].plot(y,x)
plt.tight_layout
plt.show()
```

```python
grafico = plt.figure()
ejes = grafico.add_axes([0,0,1,1])
ejes.plot(x, 2*x, label = '2*x', color = '#ff00ff', alpha=0.4, linestyle='dashed')
ejes.plot(x, x+2, label = '2+x', linewidth=0.5, linestyle='dotted')
ejes.legend()
plt.show()
```

```python
#EXPORTAR LAS IMÁGENES
```

```python
grafico.savefig('migrafico.jpg', dpi = 1080)
```
