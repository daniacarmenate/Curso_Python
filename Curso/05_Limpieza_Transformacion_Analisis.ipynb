### Tratamiento de duplicados

```python
# Función drop_duplicates, evalua el dataset y elimina las filas duplicadas
salaries.drop_duplicates()
```

### Tratamiento de valores nulos

```python
# Opción 1: Elimar todos los valores nulos:

# Eliminar filas que contienen al menos un valor nulo
df_sin_nulos = salaries.dropna()

# Eliminar columnas que contienen al menos un valor nulo
df_sin_nulos_columnas = salaries.dropna(axis=1)
```

```python
df_sin_nulos.shape
```

```python
df_sin_nulos_columnas.shape
```

```python
# Opción 2: Reemplazar por la media
df_sin_nulos["salary"]
```

```python
# La columna salary contiene valores de salario pero formato de cadena
# eliminar cualquier valor no numérico, extraer los digitos y convertir la columna resultante en valores tipo float

df_sin_nulos['salary_numeric'] = df_sin_nulos['salary'].replace('[^\d.]','',regex=True).str.extract('(\d+)').astype(float)
```

```python
df_sin_nulos['salary_numeric']
```

```python
df_sin_nulos['salary_numeric'].describe()
```

```python
# Crear un boxplot
plt.boxplot(df_sin_nulos['salary_numeric'])

# Añadir etiquetas y título
plt.xlabel("Salarios")
plt.ylabel("Valores en USD")
plt.title("Boxplot")

plt.show()
```

```python
# Crear un histograma
plt.hist(df_sin_nulos['salary_numeric'], bins=30, color='blue', alpha=0.7)

# Añadir etiquetas y título
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title("Histograma")

plt.show()
```

```python
## Eliminamos las filas que contengan un salario  menos a 20000
df_sin_nulos=df_sin_nulos[df_sin_nulos['salary_numeric']>=20000]
```

```python
df_sin_nulos
```

### Método de reemplazo de valores nulos con la media de la columna

```python
# Rellenar valores nulos con la mdia de la columna
df_con_relleno = df_sin_nulos.fillna(df_sin_nulos['salary_numeric'].mean())
```

### Eliminar una columna

```python
df_con_relleno.drop('salary',axis=1, inplace=True)
```

### Agrupar y agregar datos

```python
# Supongamos que tienes un dataframe llamado df con culomnas 'title' y 'salary'
# Agregar port titulo y calcular el promedio y la mediana del salario de cada titulo
df_grouped = df_con_relleno.groupby('title')['salary_numeric'].agg(['mean', 'median', 'count']).reset_index()

# Renombrar las columnas para mayor claridad
df_grouped.columns = ['title', 'salary_mean', 'salary_median', 'salary_count']

df_grouped
```

```python
# Redondear
df_grouped['salary_mean'] = df_grouped['salary_mean'].round(0).astype(int)
df_grouped
```

### Export a CSV

```python
# Exportar a CSV
# Especificar ña ruta y el nombre
ruta_nombre = "C:/Users/dania/Documents/Python Scripts/Ficheros/export_linkedin-jobs-usa.csv"

df_grouped.to_csv(ruta_nombre)
```
